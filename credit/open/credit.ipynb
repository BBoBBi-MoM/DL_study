```python
credit_df = raw_data
```

```python
credit_df['DAYS_EMPLOYED'][credit_df['DAYS_EMPLOYED']>0] =0

credit_df.iloc[:,1] = credit_df.iloc[:,1].map({'M':0,'F':1})
credit_df.iloc[:,2] = credit_df.iloc[:,2].map({'N':0,'Y':1})
credit_df.iloc[:,3] = credit_df.iloc[:,3].map({'N':0,'Y':1})

income_one_hot = pd.get_dummies(credit_df.iloc[:,6])
edu_one_hot = pd.get_dummies(credit_df.iloc[:,7])
family_one_hot = pd.get_dummies(credit_df.iloc[:,8])
house_one_hot = pd.get_dummies(credit_df.iloc[:,9])
occyp_one_hot = pd.get_dummies(credit_df.iloc[:,-4])

credit_df = credit_df.drop(['index','income_type','edu_type','family_type','house_type','occyp_type','begin_month'], axis=1)

credit_df = pd.concat([credit_df,income_one_hot,edu_one_hot,family_one_hot,house_one_hot,occyp_one_hot],axis=1)
credit_df
```

민맥스로 정규화하기..

```python
min_max_df = credit_df
min_max_df['DAYS_BIRTH'] = -1*credit_df['DAYS_BIRTH']
min_max_df['DAYS_EMPLOYED'] = -1*credit_df['DAYS_EMPLOYED']
min_max_df['income_total'] = (credit_df['income_total'] - min(credit_df['income_total']))/(max(credit_df['income_total'])-min(credit_df['income_total']))
min_max_df['DAYS_BIRTH'] = (credit_df['DAYS_BIRTH'] - min(credit_df['DAYS_BIRTH']))/(max(credit_df['DAYS_BIRTH'])-min(credit_df['DAYS_BIRTH']))
min_max_df['DAYS_EMPLOYED'] = (credit_df['DAYS_EMPLOYED'] - min(credit_df['DAYS_EMPLOYED']))/(max(credit_df['DAYS_EMPLOYED'])-min(credit_df['DAYS_EMPLOYED']))
```

```python
min_max_df.to_csv('min_max.csv', index=False)
min_max_df
```
